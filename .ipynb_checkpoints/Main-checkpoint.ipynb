## Data prepossessing

Load raw data

```python
import os
import pandas as pd
pd.options.display.max_columns = None

class LocationData:
  def __init__(self, folder_path: str):
    self.test_est = self.read_parquet(folder_path, "X_test_estimated.parquet")
    self.train_est = self.read_parquet(folder_path, "X_train_estimated.parquet")
    self.train_obs = self.read_parquet(folder_path, "X_train_observed.parquet")
    self.train_tar = self.read_parquet(folder_path, "train_targets.parquet")

  def read_parquet(self, folder_path: str, file_path: str) -> pd.DataFrame:
    return pd.read_parquet(os.path.join(folder_path, file_path))

class Data:
  def __init__(self):
    self.a = LocationData("/content/data/A")
    self.b = LocationData("/content/data/B")
    self.c = LocationData("/content/data/C")

raw_data = Data()
```
